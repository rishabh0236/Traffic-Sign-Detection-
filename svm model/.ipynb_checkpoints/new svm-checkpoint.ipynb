```python
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
import random
```

```python
dataPath = r'D:\books\4th year 2nd sem books\bs project\dataset\data\dataset'
```

```python
def loadDataSet(dataPath):
    data = []
    imageSize = 28
    for cat in os.listdir(dataPath):
        label = int(cat[:2])
        for i in os.listdir(dataPath+'/'+cat):
            imagePath = dataPath+'/'+cat+'/'+i
            try:
                image = cv2.imread(imagePath,0) 
                image = cv2.resize(image,(imageSize,imageSize))
                data.append([image,label])
            except:
                pass
    
    random.shuffle(data)
    
    X = [] #image
    Y = [] #labels
    for img,l in data: 
        X.append(img)
        Y.append(l)
    X = np.array(X)
    Y = np.array(Y).reshape(len(Y),1)
    return X,Y
```

```python
#Load Data Set 
images,labels=loadDataSet(dataPath)
```

```python
#Splitting Data into Training Set and Testing Set
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split


x_train,x_test,y_train,y_test=train_test_split(images,labels,test_size=0.3,random_state=42)
print("the Number of Training Example=",x_train.shape[0])
print("the Number of Testing Example=",x_test.shape[0])
```

```python
def preprocessingImage(image=None,imageSize=28,mu=122.10301047296389,std=57.11581528540206):
    try:
        image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    except:
        pass
    image = cv2.resize(image,(imageSize,imageSize))
    image = (image - mu) / std
    image = image.reshape(1,imageSize,imageSize,1)
    return image
```

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# scalling the input data
sc_X = StandardScaler() 
x_train = sc_X.fit_transform(x_train.reshape(-1, x_train.shape[-1]))
x_test = sc_X.fit_transform(x_test.reshape(x_test.shape[0], -1))
y_train = sc_X.fit_transform(y_train.reshape(-1, y_train.shape[-1]))
y_test = sc_X.fit_transform(y_test.reshape(y_test.shape[0], -1))
```

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
# importing SVM module
from sklearn.svm import SVC

# kernel to be set linear as it is binary class
classifier = SVC(kernel='linear')

# traininf the model
classifier.fit(x_train,y_train)
```

```python
x_test = x_test[:, :28]
```

```python
# testing the model
y_pred = classifier.predict(x_test)
```
